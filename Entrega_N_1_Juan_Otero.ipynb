```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/Juanotero2811/DataScience/refs/heads/main/golf_brand_sales_us.csv', sep=',')

suma_por_categoria = df.groupby('Brand')['TotalPrice'].sum()
suma_por_categoria.plot(kind='bar')
plt.xlabel('Marca')
plt.ylabel('Total de Ventas')
plt.title('Total de Ventas por Marca de Producto')
plt.xticks(rotation=45, ha='right') # Rotar 45 grados, alinear a la derecha
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/Juanotero2811/DataScience/refs/heads/main/golf_brand_sales_us.csv', sep=',')

suma_por_categoria = df.groupby('ProductName')['TotalPrice'].sum()
suma_por_categoria.plot(kind='barh')
plt.xlabel('Total de Ventas')
plt.ylabel('Categoría de productos')
plt.title('Total de Ventas por Categoría de Producto')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/Juanotero2811/DataScience/refs/heads/main/golf_brand_sales_us.csv', sep=',')

suma_por_categoria = df.groupby('SalesChannel')['TotalPrice'].sum()
suma_por_categoria.plot(kind='pie')
plt.xlabel('Ventas')
plt.ylabel('Canal')
plt.title('Total de Ventas por canal')
plt.show()
```
